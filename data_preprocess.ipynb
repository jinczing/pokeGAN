```python
import torch
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, transform, color
from sklearn.preprocessing import PowerTransformer
from bitarray import bitarray
from bitarray.util import int2ba
```

```python
# image resize and convert channel
raw_image_dir = './pokemon_images/full/'
new_image_dir = './pokemon_images/pre/'

for image in os.listdir(raw_image_dir):
    img = io.imread(os.path.join(raw_image_dir, image))
    img = transform.resize(img, (128, 128, 3))
    #img = color.rgba2rgb(img, background=(0, 0, 0))
    io.imsave(os.path.join(new_image_dir, image), img)
```

```python
# standarize attributes
def standarize_and_clip(x):
    return (np.clip(((x-np.mean(x)))/np.std(x), -2, 2)+2)/4
    # return ((x-np.mean(x)))/np.std(x)

pt_h = PowerTransformer()
pt_w = PowerTransformer()
pt_h.fit(attribute_data[:, 10].reshape(-1, 1))
pt_w.fit(attribute_data[:, 11].reshape(-1, 1))
attribute_data[:, 10] = standarize_and_clip(pt_h.transform(attribute_data[:, 10].reshape(-1, 1)).reshape(-1))
attribute_data[:, 11] = standarize_and_clip(pt_w.transform(attribute_data[:, 11].reshape(-1, 1)).reshape(-1))
for i in range(12, 18):
    attribute_data[:, i] = standarize_and_clip(attribute_data[:, i])

plt.figure(figsize=(8,16))
for i in range(10, 18):
    print(np.mean(attribute_data[:,i]), np.std(attribute_data[:,i]))
    plt.subplot(8, 1, i-9)
    plt.hist(attribute_data[:,i], n_bins)
```

```python
# write as csv
with open('./attributes.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(attribute_data)
```
